```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
b1 = pd.read_csv(r'C:\Users\tonya\Downloads\baseball\BASEBALL_1M.csv', index_col = 0)
```

```python
b1
```

```python
b1sn = nx.Graph(b1)
```

```python
print(nx.info(b1sn))
```

```python
nx.average_clustering(b1sn)
```

Average clustering, or the tendency of this social network to form clusters expressed as a fraction (in this case, represented as a decimal) is 0.9117143658810325.

```python
nx.betweenness_centrality(b1sn)
```

The only players that have reports of betweenness centrality, characterized as the likeliness that others will pass through that particular player, are Jason Giambi, Gary Sheffield, Barry Bonds, Armando Rios, Jason Grimsley, Gary Matthews Jr, David Bell, and Jose Canseco. The player with the highest betweenness centrality is Jerry Harrison Jr at 0.017303822937625755.

```python
nx.degree_centrality(b1sn)
```

Jay Gibbons, Rick Ankiel, Troy Glaus, and Scott Schoeneweis all have the lowest degree centrality at 0.04225352112676056.

```python
plt.figure(3,figsize=(12,12))
nx.spring_layout(b1sn)
nx.draw_networkx(b1sn, node_size = 10)
```

```python
b1di = nx.DiGraph(b1)
```

```python
plt.figure(3,figsize=(12,12))
nx.spring_layout(b1di)
nx.draw_networkx(b1di)
```

Although the 4 players previously mentioned have the same degree centrality, Scott Schoeneweis and Troy Glaus only connect to each other, forming their own little connection. I would like to see if some of the results that I have already uncovered, such as the clustering coefficient, degree centrality and betweenness centrality will differ Schoeneweis and Glaus are removed from the dataset.

```python
b1new = b1.drop(columns = ['Scott_Schoeneweis', 'Troy_Glaus']) #drop them from columns
```

```python
b1update = b1new.drop(b1new.index[17])
```

```python
b1final = b1update.drop(b1update.index[16])
```

Using the drop() function, Scott Schoeneweis and Troy Glaus have both been removed from the columns and rows of this dataframe.

```python
#b1final.to_csv(r'C:\Users\tonya\Downloads\baseball\b1final.csv')
```

```python
final_sn = nx.Graph(b1final)
```

```python
print(nx.info(final_sn))
```

The amount of nodes has gone from 72 to 70. The amount of edges has gone from 1089 to 1086.

```python
nx.average_clustering(final_sn)
```

The clustering coefficient has increased. Before the removal of Schoeneweis and Glaus, the clustering coefficient was 0.9117143658810325. Now, it is 0.9377633477633477.

```python
nx.betweenness_centrality(final_sn)
```

The players who have a value for betweenness centrality has stayed the same, but the values for each of these players has increased by a small fraction.

```python
nx.degree_centrality(final_sn)
```

```python
plt.figure(3,figsize=(12,12))
nx.spring_layout(final_sn)
nx.draw_networkx(final_sn, node_size = 10)
```

```python
final_di = nx.DiGraph(b1final)
```

```python
plt.figure(3,figsize=(12,12))
nx.spring_layout(final_di)
nx.draw_networkx(final_di)
```

Removing Schoeneweis and Glaus did not have much effect on the graph and who is connected to who, but having a smaller number (albeit just 2 less than the original) of nodes has slightly increased the strength between the nodes that are still present here.
